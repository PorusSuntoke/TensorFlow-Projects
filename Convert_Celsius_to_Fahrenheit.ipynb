```
sns.scatterplot(Temperature_df['Celsius'], Temperature_df['Fahrenheit'])
```

STEP #3: CREATE TESTING AND TRAINING DATASET

```
X_train = Temperature_df['Celsius']
y_train = Temperature_df['Fahrenheit']
```

STEP #4: BUILD AND TRAIN THE MODEL

```
X_train.shape
```

```
y_train.shape
```

```
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
```

```
model.summary()
```

```
model.compile(optimizer = tf.keras.optimizers.Adam(0.9), loss = 'mean_squared_error')
```

```
epochs_hist = model.fit(X_train, y_train, epochs = 500)
```

STEP #5: EVALUATE THE MODEL

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```
model.get_weights()
```

```
Temperature_C = 100
Temperature_F = model.predict([Temperature_C])
print('Temperature in DegF using ANN =', Temperature_F) 
```

```
# Let's confirm with the equation
Temperature_F = Temperature_C * (9/5) + 32
print('Temperature in DegF using Equation =', Temperature_F)
```
